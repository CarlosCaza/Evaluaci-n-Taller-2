```python
import logging
from datetime import datetime
import os

# Configuración básica de registros (Logs) para ver el flujo en la terminal
logging.basicConfig(level=logging.INFO, format='%(asctime)s - [%(levelname)s] - %(message)s')

class HealthNewsPipeline:
    def __init__(self, storage_format="csv"):
        self.storage_format = storage_format
        self.execution_id = datetime.now().strftime("%Y%m%d_%H%M%S")
        
    def run(self):
        logging.info(f"=== Iniciando Pipeline de Desinformación en Salud [ID: {self.execution_id}] ===")
        
        # ------------------------------------------------------------------
        # PASO 1: EXTRACCIÓN (Aquí llamarás al código de la Persona 2)
        # ------------------------------------------------------------------
        logging.info("Fase 1/3: Extrayendo datos desde feeds RSS...")
        raw_data = self._extract()
        
        if not raw_data:
            logging.warning("No se obtuvieron datos. Cancelando pipeline.")
            return

        # ------------------------------------------------------------------
        # PASO 2: TRANSFORMACIÓN/LIMPIEZA (Aquí irá el código de la Persona 1)
        # ------------------------------------------------------------------
        logging.info(f"Fase 2/3: Procesando y limpiando {len(raw_data)} artículos...")
        cleaned_data = self._transform(raw_data)

        # ------------------------------------------------------------------
        # PASO 3: CARGA / ALMACENAMIENTO (Tu responsabilidad directa)
        # ------------------------------------------------------------------
        logging.info(f"Fase 3/3: Guardando datos en formato [{self.storage_format.upper()}]...")
        self._load(cleaned_data)
        
        logging.info("=== Pipeline ejecutado con éxito ===")

    def _extract(self):
        # SIMULACIÓN: Reemplazar por: from scraper import extraer_noticias
        # Retorna la estructura que acordaste con la Persona 1 y 2
        return [
            {"titulo": "Alerta OMS por nuevo brote", "fuente": "WHO News", "fecha": "2026-06-26", "texto": "<p>Texto crudo...</p>"},
            {"titulo": "Estudio de la BBC sobre vacunas", "fuente": "BBC Health", "fecha": "2026-06-25", "texto": "Texto normal"}
        ]

    def _transform(self, data):
        # SIMULACIÓN: Aquí la Persona 1 pasará esto por Pandas para quitar HTML
        return data

    def _load(self, data):
        # Llama a tu módulo de almacenamiento
        print(f"--> [OK] {len(data)} registros guardados en /data/noticias_{self.execution_id}.{self.storage_format}")

if __name__ == "__main__":
    # Al ejecutar 'python src/pipeline.py', arranca la arquitectura
    sistema = HealthNewsPipeline(storage_format="csv")
    sistema.run()
```
